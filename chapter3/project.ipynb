Projet : Jeu des devinettes sans Fonctions ni Exceptions
Objectif
Créer un jeu de devinettes interactif où l'utilisateur doit deviner un nombre secret en utilisant les concepts de contrôle de flux et de boucles en Python.

Étapes du projet

- Générer un nombre secret aléatoire entre 1 et 100.
- Initialiser un compteur de tentatives à 0.
- Utiliser une boucle while pour permettre à l'utilisateur de saisir un nombre jusqu'à ce qu'il devine le nombre secret.
- Utiliser des instructions if, elif, et else pour donner des indices à l'utilisateur.
- Utiliser un opérateur ternaire pour un message spécial.
- Afficher le nombre de tentatives nécessaires pour deviner le nombre secret.
- Demander à l'utilisateur s'il veut rejouer.

```python
# Expressions, déclarations et types d'exécution

import random

# generation d'un nombre entre 1 et 100
nbre_secret = random.randint(1, 100)

# initialisation de la variable tentative a zero
tentative = 0

difficulte = input("choissisez votre difficulte facile/difficile/moyen").lower()

match difficulte:
    case "facile":
        max_tentative = 15
    case "moyen":
        max_tentative = 10
    case _:
        max_tentative = 5

# max_tentative = 5


#boucle while

while tentative < max_tentative:
    
    # utilisation de la fonction enumerate()
    
    for index, value in enumerate(range(3), start=1):
        guess = input(f"Tentative {tentative} , essai {index/3} pour l'estimation")
        
        # utilisation des structures contionnele
        
        if not guess.isdigit():
            print("Entrer un nombre valide")
            continue
        
        guess = int(guess)
        tentative = tentative + 1
        
        # operateur logique
        
        if guess < 1 or guess > 100:
            print(" Le nombre doit etre entre 1 et 100")
        
            break
        # print(nbre_secret)
        if guess == nbre_secret:
            print(f"Bravo vous avez trouve en tentative {tentative} tentatives")
            
            break
        
        elif guess < nbre_secret:
            print("Votre nombre est tres bas")
        else:
            print("Votre nombre est tres haut")
        
        
        # intruction break
        
        if guess == nbre_secret:
            break
        
        # intuction continue
        
        if tentative < max_tentative:
            print("Nouvelle Serie d'essai")
            continue
        print("Desole, vous avez epuisez votre stock de tentative")
        
        
    #Utilisation de l'operateur ternaire
    
    resultat = "winner" if guess == nbre_secret else "Lost"
    print(resultat)
    
print("Voici le nombre que vous auriez pu deviner ", nbre_secret)
  
```
